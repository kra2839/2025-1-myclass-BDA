```python
import pandas as pd

df =pd.read_csv('/content/행정안전부_지역별(행정동) 성별 연령별 주민등록 1인세대수_20250430.csv', encoding='euc-kr')

df
```

##지역별 1인가구 남녀 수와 더 많은 성별(대구)

```python
# '대구광역시' 데이터만 필터링
df_daegu = df[df['시도명'] == '대구광역시'].copy()

# 지역 이름을 하나로 결합
df_daegu['지역'] = df_daegu['시군구명'] + ' ' + df_daegu['읍면동명']

# 필요한 열만 추출
df_daegu_subset = df_daegu[['지역', '남자', '여자']]

# 남녀 수 비교
df_daegu_subset['더 많은 성별'] = df_daegu_subset.apply(
    lambda row: '남자' if row['남자'] > row['여자'] else '여자', axis=1)

# 결과 출력 (괄호 제대로 닫음!)
print(df_daegu_subset.sort_values('지역').reset_index(drop=True))
```

```python
# '대구광역시'만 필터링
df_daegu = df[df['시도명'] == '대구광역시'].copy()

# 시군구별로 남자 + 여자를 합산하여 전체 1인 가구 수 계산
df_daegu['총 1인가구'] = df_daegu['남자'] + df_daegu['여자']

# 시군구 단위로 그룹화하여 총합 계산
gu_summary = df_daegu.groupby('시군구명')['총 1인가구'].sum().reset_index()

# 가장 많은 시군구 찾기
top_gu = gu_summary.sort_values('총 1인가구', ascending=False)

# 결과 출력
print(top_gu)
```
